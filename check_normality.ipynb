## other time points

```python
def check_normality(Eyes, Drug, TP, Band):
    if Eyes == 'Closed':
        data_name = 'EC_relPower.csv'
    elif Eyes == 'Open':
        data_name = 'EO_relPower.csv'
    else:
        raise ValueError("Wrong Eye condition")
    # load dataframe
    df = pd.read_csv(data_name)
    # drug condition
    df = df.loc[df['Drug'] == Drug]
    # time point
    df = df.loc[df['TP'] == TP]
    # time-frequency band
    if Band == 'Alpha':
        df = df.Alpha
    elif Band == 'Beta':
        df = df.Beta
    elif Band == 'Theta':
        df = df.Theta
    else:
        raise ValueError('Wrong band name')
    # drop nans
    df = df.dropna()
    # normality test
    result = scipy.stats.normaltest(df)
    # print results
    print('Eyes = %s, Drug = %d, TP = %d, Band = %s' % (Eyes, Drug, TP, Band))
    print(result)
    if result[1] < 0.05:
        print('NOT NORMAL')
    else:
        print('NORMAL')
    print('\n')
    return result
```

```python
for eyes in ['Closed', 'Open']:
    for drug in [1, 2]:
        for tp in [1, 2, 3, 4, 5, 6]:
            for band in ['Alpha', 'Beta', 'Theta']:
                norm_result = check_normality(Eyes=eyes, Drug=drug, TP=tp, Band=band)
```
